```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
matches = pd.read_csv('IPL Matches 2008-2020.csv')
balls = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
print('Priyatama is ready!')
```

```python
balls.head()
```

```python
#Total Runs scored in each inning of each match
total_score_df = balls.groupby(['id','inning']).sum()['total_runs'].reset_index()
total_score_df
```

```python
total_score_df=total_score_df[total_score_df['inning']==1]
total_score_df
```

```python
data = pd.merge(left=matches, right=total_score_df, on='id', how='right')
data.head()
```

```python
data.head(3)
```

```python
data.team1.unique()
```

```python
teams = [
    'Royal Challengers Bangalore', 'Kings XI Punjab',
       'Delhi Daredevils', 'Mumbai Indians', 'Kolkata Knight Riders',
       'Rajasthan Royals', 'Deccan Chargers', 'Chennai Super Kings',
       'Sunrisers Hyderabad','Delhi Capitals']
```

```python
data= data[data['team1'].isin(teams)]
data = data[data['team2'].isin(teams)]
```

```python
data['team1']=data['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
data['team1']=data['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
data.columns
```

```python
data.method.unique()
```

```python
data['method'] = data['method'].fillna(value=0)
```

```python
data.method.unique()
```

```python
data2 = data[data['method']==0]
data2.head()
```

```python
data3 = data2[['id','city','venue','winner','total_runs']]
```

```python
data4 = data3.merge(balls,on='id')
```

```python
data5 = data4[data4['inning']==2]
```

```python
data5.head()
```

```python
data5['current_score'] = data5.groupby('id').cumsum()['total_runs_y']
data5.head()
```

```python
data5['target'] = data5['total_runs_x']+1
data5['runs_left']=data5['target']-data5['current_score']
```

```python
data5['balls_left']=126-((data5['over']*6)+data5['ball'])
data5.head()
```

```python
data5.columns
```

```python
data5['batting_team']=data5['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
data5['bowling_team']=data5['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
data5.drop(['non_striker','batsman','bowler','batsman_runs','extra_runs','non_boundary','dismissal_kind','player_dismissed','fielder','extras_type'], axis=1, inplace=True)
data5.columns
```

```python
data5.is_wicket.value_counts()
```

```python
wickets = data5.groupby('id').cumsum()['is_wicket'].values
data5['wickets'] =10 - wickets
data5.head()
```

```python
#crr = runs/over
data5['crr']=(data5['current_score']*6)/(120-data5['balls_left'])
```

```python
#rrr = runs_left/overs_left
data5['rrr'] = (data5['runs_left']*6)/(data5['balls_left'])
```

```python
data6 = data5.copy()
data6.head(3)
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
data6['result']=data6.apply(result, axis=1)
data6.head(3)
```

```python
data6.winner.unique()
```

```python
data6.batting_team.unique()
```

```python
data6['winner']=data6['winner'].str.replace('Delhi Daredevils','Delhi Capitals')
data6['winner']=data6['winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
data6['batting_team']=data6['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
data6['batting_team']=data6['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
data6[data6['id']==1237181].head(3)
```

```python
data6.drop(['winner','venue'], axis=1, inplace=True)
data6.columns
```

```python
final_df = data6[['batting_team','bowling_team','city', 'runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.head(3)
```

```python
final_df = data6.sample(data6.shape[0])
final_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
final_df.dropna(inplace=True)
final_df = final_df[final_df['balls_left']!=0]
```

```python
final_df = final_df.reset_index()
```

```python
final_df.columns
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 1)
```

```python
x_train.head(3)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_features = ['batting_team','bowling_team','city']

one_hot=OneHotEncoder()
trf = ColumnTransformer([("one_hot", one_hot,categorical_features )],remainder = "passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
x_train.head(3)
```

```python
y_train.head(3)
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
